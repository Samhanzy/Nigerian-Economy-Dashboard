```python
import pandas as pd
```

```python
df = pd.read_csv('../data/nigeria_macro_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Inflation_Rate', data=df, marker='o', linewidth=2)
plt.title(' Inflation Rate Over Time (Nigeria)', fontsize=14)
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Exchange_Rate_USD', data=df, marker='o', color='orange', linewidth=2)
plt.title('USD to NGN Exchange Rate Over Time', fontsize=14)
plt.ylabel('NGN per USD')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='GDP_Growth', data=df, marker='o', color='green', linewidth=2)
plt.axhline(0, linestyle='--', color='red', label='Recession Threshold')
plt.title('Nigeria GDP Growth Over Time (%)', fontsize=14)
plt.ylabel('Annual GDP Growth (%)')
plt.xlabel('Year')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Unemployment_Rate', data=df, label='Unemployment Rate', marker='o')
sns.lineplot(x='Year', y='Inflation_Rate', data=df, label='Inflation Rate', marker='o')
plt.title('Unemployment vs Inflation in Nigeria', fontsize=14)
plt.ylabel('Percentage (%)')
plt.xlabel('Year')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
correlation_matrix = df.drop(columns='Year').corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Nigerias Economic Indicators', fontsize=14)
plt.tight_layout()
plt.show()
```
